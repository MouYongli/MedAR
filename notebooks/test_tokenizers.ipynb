```python
from transformers import BertTokenizerFast

# 初始化一个BERT tokenizer
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')

# 示例句子和单词
sentence = "This is an example sentence where we want to find the position of a word."
word = "example"

# 对句子进行编码
encoded_input = tokenizer(sentence, return_offsets_mapping=True)
tokens = tokenizer.convert_ids_to_tokens(encoded_input["input_ids"])
offset_mapping = encoded_input["offset_mapping"]

# 找到原单词在tokens中的位置
original_position = sentence.split().index(word)  # 假设单词不会重复
print(original_position)
token_position = None
for i, (start, end) in enumerate(offset_mapping):
    if start <= len(" ".join(sentence.split()[:original_position])) < end:
        token_position = i
        break

tokens, token_position
```
